### **5) Estatística Descritiva Básica das colunas numéricas:**

Para a exibição somente dos dados numéricos (**Quantidade de Vendas** e **Preço Unitário**) coloquei uma função para **não** exibir a coluna de identificação (ID).

```python
df = df.drop(df.columns[0], axis=1)
df.describe()
```

### **6) Visualização de Dados:** 

#### 6.1) Gráfico de Barras

Definindo os **eixos x** como **categoria do curso** e **eixo y** como **contagem**. 

Feito isso, podemos observar a contagem dos cursos a partir do gráfico de barras abaixo.

```python
eixo_x = df['Nome do Curso'].value_counts().keys()
eixo_y = df['Nome do Curso'].value_counts().values
```

```python
plt.bar(eixo_x, eixo_y)

plt.title('Contagem de Categoria')

plt.xlabel('Nome do Curso')
plt.ylabel('Contagem')

plt.xticks(rotation=90)

plt.savefig('contagem_categoria.pdf', format = 'pdf', bbox_inches = 'tight')
plt.show()
```

#### 6.2) Gráfico de Dispersão

Definindo os **eixos x** do DataFrame como **Quantidade de Vendas** e **eixo y** do DataFrame como **Preço Unitário**. 

Feito isso, podemos observar pelo gráfico de dispersão que quanto maior o preço do curso, menor é a quantidade de curso vendido, por exemplo, o curso com o valor de 120,00 reais vendeu poucas unidades em relação ao curso com o valor de 40,00 reais que vendeu entre 45 e 50.

```python
plt.scatter(df['Quantidade de Vendas'], df['Preço Unitário'])

plt.title('Vendas em relação ao Valor')
plt.xlabel('Quantidade de Vendas')
plt.ylabel('Preço Unitário')

plt.savefig('vendas_valor.pdf', format = 'pdf', bbox_inches = 'tight')
plt.show()
```

### **8) Distribuição das Vendas ao longo do tempo**

Escolhi dta como a variável de Data e qtd como a Quantidade de Vendas do curso. Analisando o gráfico abaixo podemos perceber que no inicío Janeiro de 2023 entre os dias 01 e 04 e quase final entre os dias 18 à 21 aproximadamente, teve uma alta na quantidade de cursos vendidos, já entre os dias 05 à 17 e 23 à 31 aproximadamente, houve uma queda na quantidade de vendas.

```python
df['Data'] = pd.to_datetime(df['Data'])

dta = df['Data']
qtd = df['Quantidade de Vendas']

plt.title('Vendas ao longo do Tempo')

plt.xlabel('Datas')
plt.ylabel('Vendas')
plt.xticks(rotation=45)

plt.plot(dta, qtd, linewidth = 2)

plt.savefig('vendas_tempo.pdf', format = 'pdf', bbox_inches = 'tight')
plt.show()
```
